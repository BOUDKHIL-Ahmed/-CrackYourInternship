```python
import numpy as np

# Fonction d'activation (sigmoïde) et sa dérivée
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)  # Dérivée de la sigmoïde

# Initialisation des données (exemple XOR)
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

y = np.array([[0], [1], [1], [0]])  # Sorties attendues

# Initialisation des paramètres du réseau
np.random.seed(1)
input_neurons = 2  # Nombre de neurones d'entrée
hidden_neurons = 2  # Nombre de neurones dans la couche cachée
output_neurons = 1  # Nombre de neurones de sortie

# Poids aléatoires entre -1 et 1
weights_input_hidden = np.random.uniform(-1, 1, (input_neurons, hidden_neurons))
weights_hidden_output = np.random.uniform(-1, 1, (hidden_neurons, output_neurons))

# Bias aléatoires
bias_hidden = np.random.uniform(-1, 1, (1, hidden_neurons))
bias_output = np.random.uniform(-1, 1, (1, output_neurons))

# Taux d'apprentissage
learning_rate = 0.5

# Entraînement du réseau
epochs = 10000  # Nombre d'itérations

for epoch in range(epochs):
    # --- Propagation avant ---
    hidden_input = np.dot(X, weights_input_hidden) + bias_hidden
    hidden_output = sigmoid(hidden_input)

    final_input = np.dot(hidden_output, weights_hidden_output) + bias_output
    final_output = sigmoid(final_input)

    # --- Calcul de l'erreur ---
    error = y - final_output

    # --- Propagation arrière ---
    output_gradient = error * sigmoid_derivative(final_output)
    hidden_error = output_gradient.dot(weights_hidden_output.T)
    hidden_gradient = hidden_error * sigmoid_derivative(hidden_output)

    # Mise à jour des poids et biais
    weights_hidden_output += hidden_output.T.dot(output_gradient) * learning_rate
    bias_output += np.sum(output_gradient, axis=0, keepdims=True) * learning_rate

    weights_input_hidden += X.T.dot(hidden_gradient) * learning_rate
    bias_hidden += np.sum(hidden_gradient, axis=0, keepdims=True) * learning_rate

    # Affichage de l'erreur toutes les 1000 époques
    if epoch % 1000 == 0:
        print(f"Epoch {epoch}, Erreur: {np.mean(np.abs(error))}")

# --- Test du réseau après entraînement ---
print("\nRésultats après entraînement:")
print(final_output)

```
